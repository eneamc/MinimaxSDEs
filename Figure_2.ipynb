```python
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams.update({'font.size': 14})
plt.rcParams["figure.figsize"] = (7,5)
import seaborn as sns
colors = sns.color_palette("colorblind")

from pathlib import Path

import sys
quality = 1000
```

```python
N = 10
ld = 1
lu = 1
rho_pos_small = 1
rho_pos_extreme = 10
rho_neg= -1

D =25
```

```python
x,y = np.meshgrid(np.linspace(-ld,lu,N),np.linspace(-ld,lu,N))
```

```python
# Set the size of the figure

u = -y
v = x

# Adjust the color, scale, and alpha of the arrows
plt.quiver(x, y, u, v, color=colors[0], scale=11, alpha=1)

# Add grid lines
plt.grid(True, linestyle='--', alpha=0.5)


plt.title(r'Vector Field: $-F(x,y)$',fontsize=D)
plt.xlabel(r'$x$',fontsize=D)
plt.ylabel(r'$y$',fontsize=D)
plt.xticks(fontsize=17)
plt.yticks(fontsize=17)

# Customize the axis limits for a better view
plt.xlim(-1.2, 1.2)
plt.ylim(-1.2, 1.2)

plt.show()
```

```python
# Set the size of the figure

u = -y - rho_pos_small*x
v = +x - rho_pos_small*y

# Adjust the color, scale, and alpha of the arrows
plt.quiver(x, y, u, v, color=colors[0], scale=11, alpha=1)


# Add grid lines
plt.grid(True, linestyle='--', alpha=0.5)

plt.title(r'Vector Field: $-F + \rho \nabla F F$ (Small $\rho$)',fontsize=D)
plt.xlabel(r'$x$',fontsize=D)
plt.ylabel(r'$y$',fontsize=D)
plt.xticks(fontsize=17)
plt.yticks(fontsize=17)

# Customize the axis limits for a better view
plt.xlim(-1.2, 1.2)
plt.ylim(-1.2, 1.2)

plt.show()
```
